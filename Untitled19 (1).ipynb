```r
#Question 2
#a
id<-c(1,2,3,4,5)
name<-c("Peter","Amy","Ryan","Gary","Michelle")
salary<-c(623.30,515.20,611.00,729.00,843.25)
df<-data.frame(id,name,salary)
#b
department<-c("IT","Finance","Computer Science","Engineering","Architecture")
df$department<-department
#c
df[c(1,3,5),c(2,3)]



```

```r
#d
library(ggplot2)
subset_df<-df[c(1,4,5),]
ggplot(subset_df,aes(x=name,y=salary))+ 
  geom_bar(stat="identity")+
  labs(title="Salary of Selected Employees",x="name",y="salary")
```

```r
#e
max_salary<-min(df$salary)
min_salary<-max(df$salary)
median_salary<-median(df$salary)
salary_summary<-c(max_salary,min_salary,median_salary)
names(salary_summary)<-c("Highest","Lowest","Median")
pie(salary_summary,main="Salary Summary")
```
